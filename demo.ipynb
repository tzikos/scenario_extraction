```python
import PyPDF2

# Open the PDF file in read-binary mode
with open('scenario_extraction/doc.pdf', 'rb') as pdf_file:
    # Create a PdfFileReader object to read the PDF
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Get the number of pages in the PDF
    num_pages = len(pdf_reader.pages)
    text_list = []
    # Loop through each page in the PDF and extract its text
    for page_num in range(num_pages):
        # Get the page object
        page = pdf_reader.pages[page_num]

        # Extract the text from the page
        text_list.append(page.extract_text())

        # Print the text to the console
        text = ' '.join(text_list).replace('\n',' ')
```

```python
len(text.split())
```

```python
words = text.split()

num_parts = 10

part_size = len(words) // num_parts

parts = [words[i:i+part_size] for i in range(0, len(words), part_size)]
parts_list = []
for i, part in enumerate(parts):
    parts_list.append(' '.join(part))
```
